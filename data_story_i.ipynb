```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%pylab inline
day_df = pd.read_csv('day.csv')
hour_df = pd.read_csv('hour.csv')
```

```python
day_df['cnt'].plot(lw=0, marker = '.', title = 'Time Series of Total Rentals')
```

```python
day_df['casual'].plot(lw=0, marker = '.', title = 'Time Series of Casual Rentals')
```

```python
day_df['registered'].plot(lw=0, marker = '.', title = 'Time Series of Registered Rentals')
```

```python
day_grp_mnth = day_df.groupby('mnth')
day_grp_mtth_totals = day_grp_mnth.sum()
day_grp_mtth_totals['cnt'].plot(kind = 'bar', title = 'Total Rentals by Month')
```

```python
day_grp_mtth_totals['registered'].plot(kind = 'bar', title = 'Registered Rentals by Month')
```

```python
day_grp_mtth_totals['casual'].plot(kind = 'bar', title = 'Casual Rentals by Month')
```

```python
day_grp_wkday = day_df.groupby('weekday')
day_grp_wkday_totals = day_grp_wkday.sum()
```

```python
day_grp_wkday_totals['cnt'].plot(kind = 'bar', title = 'Total Rentals by Weekday')
```

```python
day_grp_wkday_totals['registered'].plot(kind = 'bar', title = 'Registered Rentals by Weekday')
```

```python
day_grp_wkday_totals['casual'].plot(kind = 'bar', title = 'Casual Rentals by Weekday')
```

```python
day_grp_season = day_df.groupby('season')
day_grp_season_totals = day_grp_season.sum()
```

```python
day_grp_season_totals['cnt'].plot(kind = 'bar', title = 'Total Rentals by Season')
```

```python
hour_grp_hour = hour_df.groupby('hr')
hour_grp_hour_totals = hour_grp_hour.sum()
hour_grp_hour_totals['cnt'].plot(kind = 'bar', title = 'Total Rentals by Hour')
```

```python
day_grp_weather = day_df.groupby('weathersit')
day_grp_weather_totals = day_grp_weather.sum()
day_grp_weather_totals['cnt'].plot(kind = 'bar', title = 'Total Rentals by Weather')
```

The above plots show data from 2011 and 2012 for a bike sharing system in Porto, Portugal.

There seems to be a year to year increase in bike rentals ["Time Series Total of Rentals"], with a sesonal low during winter increasing to a seasonal high during summer ["Total Rentals by Season"]. Total rentals are fairly level through each weekday ["Total Rentals by Weekday"], with casual rentals showing a marked increase during weekends ("Casual Rentals by Weekday"), and registered rentals showing a bit of the opposite (lower during weekends) ("Registered Rentals by Weekday").

By hour, peak rentals occur in the morning and early evening ("Total Rentals by Hour"), which coincides with the hours people usually leave for work and return home from work. From the "Total Rentals by Weather" plot, it's clear that people rarely rent bikes when the weather has any precipitation (label 3).

It would be interesting to further investigate similarities and differences between the rental patterns of casual and registered users. A breakdown of this relationship along other lines (holidays, working days, temperature/humidity/wind speed (i.e. some concept of pleasant/unpleasant weather) should reveal other useful insights.
